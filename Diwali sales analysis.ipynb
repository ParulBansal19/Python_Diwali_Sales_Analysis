# Exploratory Data Analysis

### Gender

```python
gen= sns.countplot(x='Gender',data= dt)

for bars in gen.containers:
    gen.bar_label(bars)
```

```python
sales_gen= dt.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False)

sns.barplot(data= sales_gen, x='Gender', y= 'Amount')
```

From above graphs we can see that most of buyers are females and even the purchasing power of females more than males

### Age

```python
ag= sns.countplot(x= 'Age Group', data= dt, hue= 'Gender')

for bars in ag.containers:
    ag.bar_label(bars)
```

```python
sales_ag= dt.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.barplot(data= sales_ag, x= 'Age Group', y= 'Amount')
```

From above graphs we can see that the most of the buyers of age group 26-35 yrs females

### State

```python
# Totel amount/sales from top 10 states

sales_st= dt.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)


sns.set(rc= {'figure.figsize':(15,5)})
sns.barplot(x= 'State', y= 'Amount', data= sales_st)
```

```python
# Totel number of orders from top 10 states

sales_sta= dt.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(15,5)})
sns.barplot(data= sales_sta, x= 'State', y= 'Orders')
```

From above graphs we can see that unexpectedly most of the orders from the Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/amount is from Uttar Pradesh, Maharashtra and than Karnataka

### Marital Status

```python
ma= sns.countplot(x= 'Marital_Status', data= dt)

sns.set(rc= {'figure.figsize': (7,5)})
for bar in ma.containers:
    ma.bar_label(bars)
```

```python
sales_ma= dt.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.set({'figure.figsize': (6,4)})
sns.barplot(x= 'Marital_Status', y= 'Amount', data= sales_ma, hue= 'Gender')
```

From above graphs we can see that the most of the buyers are married (females) and they have high purchasing power

### Occuption

```python
sns.set(rc= {'figure.figsize': (20,5)})
occ= sns.countplot(x= 'Occupation', data= dt)

for bar in occ.containers:
    occ.bar_label(bars)
```

```python
sales_occ= dt.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.barplot(x= 'Occupation', y= 'Amount', data= sales_occ)
```

From above graphs we can see that most of the buyers are working in IT Sectors, Healthcare and Aviation sector

### Product Category

```python
sns.set(rc= {'figure.figsize':(25,5)})
pr= sns.countplot(x= 'Product_Category', data= dt)
```

```python
sales_pr= dt.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize': (35,5)})
sns.barplot(x= 'Product_Category', y= 'Amount', data= sales_pr)
```

From above graphs we can see that most of the sold products are Food, Clothing & Apparel and Electronics & Gadgets

```python
# Top 10 most sold products

sales_id= dt.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(x= 'Product_ID', y='Orders', data= sales_id)
```

### Summary

Married woman age group 25-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation sectors are more likely to buy products from Food, Clothing and Electronics category.
